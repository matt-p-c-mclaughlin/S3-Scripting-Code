## This file will clean the folders remaining in 10_09_2019 after I finish undeleting everything

```python
import boto3
from datetime import datetime
```

```python
s3 = boto3.resource('s3')
BUCKET_NAME = "bucknellrobotics"
IMG_BUCKET = s3.Bucket(name=BUCKET_NAME)
```

```python
def clean_folders(obj):
    key = obj.key
    split_key = key.split("/")
    
    #We could add more conditions but this sums it up:
    #I want to remove every subfolder in 10_09_2019
    if ((split_key[0] == "10_09_2019") and (len(split_key) >= 3)):
        #Then the key is of form
        #10_09_2019/ASIN/maybe other subfolders/original_key.jpg
        #and we can dump it
        #print("DELETING: " + str(obj.key))
        obj.delete()
        
```

```python
start_time = datetime.now()
i = 0
for obj in IMG_BUCKET.objects.all():
    clean_folders(obj)
    
    #Test on subset first
    #Make sure it deletes folders
    #   It does
    #Make sure it doesn't delete images
    #   Run for 212 to have it check first non-folder image
    #   It works
    #i = i +1
    #if i==214:
        #break
    
    i = (i+1)%10000
    if i==0:
        print(str(datetime.now()) + "   10,000 unecessary files deleted")
end_time = datetime.now()
print('\n\nDONE')
print("Took: " + str(end_time - start_time) + " seconds")
```
